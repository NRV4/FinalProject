```python
def score_emp(x):
    if x== '< 1 year':
        return 0
    if x=='1 year':
        return 1
    if x == '2 years':
        return 2
    if x == '3 years':
        return 3
    if x=='4 years':
        return 4
    if x=='5 years':
        return 5
    if x == '6 years':
        return 6
    if x == '7 years':
        return 7
    if x == '8 years':
        return 8
    if x == '9 years':
        return 9
    if x == '10+ years':
        return 10
```

```python
df2nm['emp_length2'] = df2nm['emp_length'].apply(score_emp)
df2nm.tail(10)
```

```python
pd.value_counts(df2nm['emp_length2'])
```

```python
import numpy as np
df2nm['term2'] = np.where(df2nm['term']==' 36 months', 36, 60)
df2nm.head(10)
```

```python
pd.value_counts(df2nm['term2'])
```

```python
def score_grade(x):
    if x == 'A':
        return 1
    if x=='B':
        return 2
    if x == 'C':
        return 3
    if x == 'D':
        return 4
    if x == 'E':
        return 5
    if x == 'F':
        return 6
    if x == 'G':
        return 7
```

```python
df2nm['grade2'] = df2nm['grade'].apply(score_grade)
df2nm.head(20)
```

```python
pd.value_counts(df2nm['grade2'])
```

```python
df2nm.describe()
```

```python
df2nm.corr()
```

```python
def score_lstatus(x):
    if x == 'Charged Off':
        return 0
    if x=='Fully Paid':
        return 1
    #if x == 'Default':
        #return 0
```

```python
df2nm['loan_status2'] = df2nm['loan_status'].apply(score_lstatus)
df2nm.head(20)
```

```python
pd.value_counts(df2nm['loan_status2'])
```

```python
df2nm.groupby("loan_status")["annual_inc"].mean()
```

```python
df2nm.groupby("loan_status2")["annual_inc"].mean()
```

```python
#remove NaN after recoding loan_status
df3nm = df2nm.dropna()
df3nm.head()
```

```python
df3nm.describe()
```

```python
df3nm.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ annual_inc + emp_length2 + term2 + grade2 + funded_amnt', data=df3nm).fit()
result.summary()
```

```python
#remove intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ annual_inc + emp_length2 + term2 + grade2 + funded_amnt - 1', data=df3nm).fit()
result.summary()
```

```python
#remove term2
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ annual_inc + emp_length2 + grade2 + funded_amnt - 1', data=df3nm).fit()
result.summary()
```

```python
#added term back in, removed emp_length2
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ annual_inc + term2+ grade2 + funded_amnt - 1', data=df3nm).fit()
result.summary()
```

```python
#added term back in, removed emp_length2, annual_inc
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ term2+ grade2 + funded_amnt - 1', data=df3nm).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ term2 + grade2 - 1', data=df3nm).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='loan_status2 ~ term2 - 1', data=df3nm).fit()
result.summary()
```
